4.3.1 –ó–∞–¥–∞–Ω–∏–µ. –ó–∞–¥–∞–π—Ç–µ –º–∞—Å—Å–∏–≤ —Å–ª—É—á–∞–π–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞(0; 1).
–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ —Å—Ä–µ–¥–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –¥–ª—è –º–∞—Å—Å–∏–≤–∞, —Å—Ä–∞–≤–Ω–∏—Ç–µ
—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã, –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –æ –∑–Ω–∞—á–µ–Ω–∏—è—Ö?
–ü–æ—Å—Ç—Ä–æ–π—Ç–µ —Ç–æ—á–µ—á–Ω—É—é –¥–∏–∞–≥—Ä–∞–º–º—É —Ä–∞—Å—Å–µ–Ω–∏—è –ø–æ–ª—É—á–µ–Ω–Ω–æ–≥–æ —Ä—è–¥–∞.

```python
from matplotlib import pyplot as pyplot
import numpy as np
import random

par = 7 # –ø–∞—Ä–∞–º–µ—Ç—Ä, –º–æ–∂–Ω–æ –º–µ–Ω—è—Ç—å
nums = list(range(1, par + 1))
x = list(range(par))
for i in range(par):
    x[i] = random.random() # —Ä–∞–Ω–¥–æ–º–Ω–æ–µ –≤–µ—â–µ—Å—Ç–≤. —á–∏—Å–ª–æ –æ—Ç 0 –¥–æ 1.0
av = np.average(x) # —Å—Ä–µ–¥–Ω–µ–µ
med = np.median(x) # –º–µ–¥–∏–∞–Ω–Ω–æ–µ
print(av, med)
if av > med:
    print("–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –ë–û–õ–¨–®–ï —á–µ–º –º–µ–¥–∏–∞–Ω–Ω–æ–µ")
elif av == med:
    print("–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –†–ê–í–ù–û –º–µ–¥–∏–∞–Ω–Ω–æ–º—É")
else:
    print("–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –ú–ï–ù–¨–®–ï –º–µ–¥–∏–∞–Ω–Ω–æ–≥–æ")
plt.scatter(nums, x)
```

4.3.2 –ó–∞–¥–∞–Ω–∏–µ. –î–∞–Ω–∞ –Ω–µ–∫–æ—Ç–æ—Ä–∞—è —Ñ—É–Ω–∫—Ü–∏—è f(x). –°–æ–∑–¥–∞—Ç—å –º–∞—Å—Å–∏–≤ –∏–∑ 10 –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ (ùë•, –Ω–∞–ø—Ä–∏–º–µ—Ä, –∏–∑–º–µ–Ω—è–µ—Ç—Å—è –æ—Ç 1 –¥–æ 10). –í—ã–¥–µ–ª–∏—Ç—å —Å—Ä–µ–∑ –ø–µ—Ä–≤–æ–π –ø–æ–ª–æ–≤–∏–Ω—ã –º–∞—Å—Å–∏–≤–∞ –∏ –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –º–∞—Å—Å–∏–≤–∞ ‚Äì –ª–∏–Ω–µ–π–Ω—ã–π –∏ –¥–ª—è —Å—Ä–µ–∑–∞ ‚Äì —Ç–æ—á–µ—á–Ω—ã–π.

```python
import math as m
import numpy as np
from matplotlib import pyplot as pyplot
import random

def f(x):
    return (m.sqrt(1 + m.e**m.sqrt(x) + m.cos(x)**2) 
    / m.fabs(1 - m.sin(x)**3)) + m.log(m.fabs(2*x)) # —Ñ—É–Ω–∫—Ü–∏—è

par = 10
nums = list(range(1, par + 1)) # –Ω–æ–º–µ—Ä–∞
nums_srez = list(range(1, (par // 2) + 1)) # –Ω–æ–º–µ—Ä–∞ –¥–ª—è —Å—Ä–µ–∑–∞
x = list(range(par))
srez = list(range(par // 2))
for i in range(par):
    x[i] = f(random.uniform(1, 10)) # –∑–∞–ø–æ–ª–Ω—è–µ–º –º–∞—Å—Å–∏–≤ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞–Ω–¥–æ–º–Ω—ã—Ö —á–∏—Å–µ–ª –æ—Ç 1 –¥–æ 10
    if i < par // 2:
        srez[i] = x[i] # –∑–∞–ø–æ–ª–Ω—è–µ–º —Å—Ä–µ–∑
plt.grid() # —Å–µ—Ç–∫–∞
plt.scatter(nums_srez, srez) # —Ç–æ—á–µ—á–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
plt.show() # –≤—Ç–æ—Ä–æ–π –≥—Ä–∞—Ñ–∏–∫
plt.grid()
plt.plot(nums, x) # –ª–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫
```

```python
import math
print("–í–≤–µ–¥–∏—Ç–µ x: ")
x = int(input())
print("–í–≤–µ–¥–∏—Ç–µ y: ")
y = int(input())
print("–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä –æ–ø–µ—Ä–∞—Ü–∏–∏, –≥–¥–µ + (1), -(2), *(3), /(4), e^(x+y) (5), sin(x+y) (6), cos(x+y) (7), x^y (8): ")
op = int(input())
if op == 1:
    res = x + y
elif op == 2:
    res = x - y
elif op == 3:
    res = x * y
elif op == 4:
    res = x / y
elif op == 5:
    res = math.e**(x + y)
elif op == 6:
    res = math.sin(x + y)
elif op == 7:
    res = math.cos(x + y)
elif op == 8:
    res = x**y
else:
    print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤–≤–æ–¥ –æ–ø–µ—Ä–∞—Ü–∏–∏")
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç = {res}")
```
